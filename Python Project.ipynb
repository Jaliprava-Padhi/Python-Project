```python
plt.figure(figsize=(8,5))
plt.bar(list(marvel['Estimated incidence of TB cases who are HIV-positive per 100 000 population'])[0:5],list(marvel['Year'])[0:5],color="g")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(marvel['Estimated incidence of TB cases who are HIV-positive'])[0:5],list(marvel['Estimated total population number'])[0:5],color=["red","green","blue","orange","pink"])
plt.show()
```

```python
#making a histogram for Estimated incidence of TB cases who are HIV-positive per 100 000 population, high bound
plt.figure(figsize=(5,7))
plt.hist(marvel['Estimated incidence of TB cases who are HIV-positive per 100 000 population, high bound'])
plt.title("Distributation of Estimated incidence of TB cases who are HIV-positive per 100 000 population, high bound")
plt.show()
```
